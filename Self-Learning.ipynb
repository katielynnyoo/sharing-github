```python
# pandas 활용하여 csv파일 데이터 불러오기 및 확인
# numpy where clause 활용위해 불러오기
import numpy as np
import pandas as pd
dataset = pd.read_csv("Workspace/Data_October2018.csv", encoding = "utf-8", delimiter = "|", low_memory=False)
dataset.head()
```

```python
# columns to lower case letters
dataset.columns = dataset.columns.str.lower()
```

```python
# data type 확인
dataset.dtypes
```

```python
# 컬럼별 건수 확인
groupResult = dataset.count()
groupResult
```

```python
# 컬럼중 최대건수 확인 (전체 건수 확인)
groupResult2 = groupResult.max()
groupResult2
```

```python
# 필수항목상 NULL 값을 "null"로 replace 
# 필수항목: 상차지, 하차지, 톤급(reqton), 기타 추가될 수 있음
dataset['pickuploc22'] = np.where(pd.notnull(dataset['pickuploc22']) == True, dataset['pickuploc22'], "null")
dataset['droploc22'] = np.where(pd.notnull(dataset['droploc22']) == True, dataset['droploc22'], "null")
```

```python
# 필수항목상 NULL 값을 "null"로 replace 
# 필수항목: distf_pickup, distf_drop, reqton, reqtruck
dataset['distf_pickup'] = np.where(pd.notnull(dataset['distf_pickup']) == True, dataset['distf_pickup'], "null")
dataset['distf_drop'] = np.where(pd.notnull(dataset['distf_drop']) == True, dataset['distf_drop'], "null")
dataset['reqton'] = np.where(pd.notnull(dataset['reqton']) == True, dataset['reqton'], "null")
dataset['reqtruck'] = np.where(pd.notnull(dataset['reqtruck']) == True, dataset['reqtruck'], "null")
```

```python
# 데이터셋 필터링 방법 1
# pickuploc22, droploc22, reqton, reqtruck 에 null값이 아닌 값만 저장
dataset2 = dataset.query('distf_pickup != "null" & distf_drop != "null" & reqton != "null" & reqtruck != "null" & distf_pickup != distf_drop & reqtruck != "덤프" ')
dataset2.head()

# 데이터셋 필터링 방법 2
# dataset = dataset[ (dataset.pickuploc22 != "null") & (dataset.droploc22 != "null") ] 
```

```python
# 컬럼별 null값 개수 확인
dataset2.isnull().sum().head()
# 필터 결과 확인 (null value 유무)
dataset2[(dataset2.pickuploc22 == "null") ].head() 
```

```python
# 컬럼별 null값 개수중 max 
countNull = dataset2.isnull().sum()
countNull.max()
```

```python
dataset2.head()
```

```python
# 신규데이터 group by 상하차지, 차종, 톤급별 건수
dataset3 = dataset2.groupby(["distf_pickup","distf_drop", "reqtruck", "reqton"])["mdmcode2"].count()
dataset3.head()
```

```python
dataset3 = pd.DataFrame(dataset3)
dataset3.head()
```

```python
# index reset하지 않으면 위에 내용처럼 테이블형태가 없어짐 (dtypes도 한개로 return됨)
dataset4 = dataset3.reset_index()
dataset4.head()
```

```python
# datatype확인
dataset4.dtypes
```

```python
# 컬럼명 바꾸기
dataset4.rename(columns={'distf_pickup': 'pick', 'distf_drop': 'drop', 'mdmcode2': 'count'}, inplace=True)
dataset4.head()
```
